```python
from dask.distributed import Client
```

```python
client = Client()
client
```

### 简单的例子

```python
def square(x):
    return x ** 2
```

```python
ret = client.map(square, [0, 1, 2, 3, 4])
```

```python
client.gather(ret)
```
